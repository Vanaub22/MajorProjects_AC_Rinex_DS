```python
import pandas as pd
df=pd.read_csv('/content/train.csv')
print(df.shape)#number of rows and columns
print(df.size)#number of elements
print(df.isnull().sum())
```

```python
df.info()#displays all relevant details
#the 177 rows missing in the age column is due to lack of information
```

```python
#to know the number of unique elements in each column
print(df.nunique())
```

```python
#find the number of people who survived(exact count)
print(df.groupby('Survived').size())
```

```python
#to find the number of males who survived the shipwreck
import numpy as np
male_surv=np.sum((df['Sex']=='male')&(df['Survived']==1))
male_dead=np.sum((df['Sex']=='male')&(df['Survived']==0))
print('Total number of males on board:',np.sum(df['Sex']=='male'))
print('The number of males who survived:',male_surv)
print('The number of males who lost their lives:',male_dead)
```

```python
import numpy as np
young=np.sum((df['Age']>=0)&(df['Age']<20))
adult=np.sum((df['Age']>=20)&(df['Age']<40))
midage=np.sum((df['Age']>=40)&(df['Age']<60))
old=np.sum(df['Age']>=60)
print('The number of young people on board:',young)
print('The number of adult people on board:',adult)
print('The number of middle-age people on board:',midage)
print('The number of old people on board:',old)
```

```python
#to find the age of the youngest and oldest passenger on board
#use loc to locate the row if required
print('Age of the youngest person on board:',np.min(df['Age']),'years')
print('Age of the eldest person on board:',np.max(df['Age']),'years')
```
